```
import tensorflow as tf
from tensorflow import keras
import numpy as np
import pandas as pd
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Flatten,Dense,Dropout,Conv1D,MaxPooling1D,BatchNormalization




data=pd.read_csv('/content/drive/My Drive/data_deep_learning/creditcard.csv')
#print(data)
#data.isnull().sum()
#data['Class'].value_counts()
nonFraud=data[data['Class']==0]
fraud=data[data['Class']==1]
nonFraud=nonFraud.sample(800)
nonFraud.shape,fraud.shape

data=fraud.append(nonFraud,ignore_index=True)
#data=tf.shuffle(data)

#print(data)

x=data.drop('Class',axis=1)
y=data['Class']
#print(y)
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
#x_train.shape,x_test.shape

scale=StandardScaler()
x_train=scale.fit_transform(x_train)

x_test=scale.transform(x_test)

x_train=x_train.reshape(1033,30,1)
x_test=x_test.reshape(259,30,1)
x_train.shape,x_test.shape

model=Sequential()
model.add(Conv1D(filters=32,kernel_size=3,activation='relu',input_shape=x_train[0].shape))
model.add(Dropout(0.5))
model.add(BatchNormalization())

model.add(Conv1D(filters=64,kernel_size=3,activation='relu',input_shape=x_train[0].shape))
model.add(Dropout(0.5))
model.add(BatchNormalization())
model.add(Conv1D(filters=128,kernel_size=3,activation='relu',input_shape=x_train[0].shape))
model.add(Dropout(0.5))
model.add(BatchNormalization())

model.add(Flatten())
model.add(Dense(128,activation='relu'))

model.add(Dense(128,activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(1,activation='sigmoid'))

model.compile(optimizer='adam',loss=tf.keras.losses.BinaryCrossentropy(),metrics=['accuracy'])
model.fit(x_train,y_train,epochs=5)
prediction=model.predict(x_test)
np.set_printoptions(suppress=True)
print(x_test[0])
print(prediction[0])
```
