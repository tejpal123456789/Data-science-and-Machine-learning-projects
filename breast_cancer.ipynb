```
import tensorflow as tf
from tensorflow import keras
import numpy as np
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Flatten,Dense,Activation,Conv1D,MaxPooling2D,Dropout,BatchNormalization
import pandas as pd
from sklearn.datasets import load_breast_cancer

cancer=load_breast_cancer()
#dataset=pd.read_csv('/content/drive/My Drive/Colab Notebooks/datasets_180_408_data.csv')
x=pd.DataFrame(data=cancer.data,columns=cancer.feature_names )
#x.head()
#df=pd.Dataframe(dataset)
#print(dataset)
y=cancer.target

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
print(x_test)
from sklearn.preprocessing import StandardScaler
scale1=StandardScaler()
x_train=scale1.fit_transform(x_train)

scale2=StandardScaler()
x_test=scale2.fit_transform(x_test)
#print(x_train[0])
#print(x_train.shape)
#print(x_test.shape)
x_train=x_train.reshape(455,30,1)
#print(x_train.shape)
x_test=x_test.reshape(114,30,1)

model=Sequential()
model.add(Conv1D(filters=32,kernel_size=2,activation='relu',input_shape=(30,1)))
model.add(BatchNormalization())
model.add(Dropout(0.2))
model.add(Conv1D(filters=32,kernel_size=2,activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.2))


model.add(Conv1D(filters=32,kernel_size=2,activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.2))
model.add(Flatten())
model.add(Dense(128,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.compile(optimizer=tf.keras.optimizers.Adam(lr=0.000005),loss=tf.keras.losses.BinaryCrossentropy(),metrics=['accuracy'])
model.fit(x_train,y_train,epochs=100)


pred=model.predict(x_test)
np.printoptions(suppress=True)
print(x_test[0])
print(pred[0])









```
